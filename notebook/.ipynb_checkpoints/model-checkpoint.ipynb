```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import joblib
```

```python
penguin = pd.read_csv("../data/penguins_cleaned.csv")
penguin.head()
```

```python
penguin.tail()
```

```python
penguin.info()
```

```python
penguin.head(2)
```

## Splitting into features and target:

```python
X = penguin.drop(columns=['species'])
y = penguin['species']
```

## Identifying categorical and numerical features:

```python
categorical_columns = X.select_dtypes(include=['object']).columns
numerical_columns = X.select_dtypes(include=['float64', 'int64']).columns
```

```python
categorical_columns
```

```python
numerical_columns
```

## Defining the transformation:

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_columns),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
])
```

## Define the full Pipeline:

```python
model = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', RandomForestClassifier(random_state=32))
])
```

## Train and Test Split:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
```

```python
model.fit(X_train, y_train)
```

## Make the prediction:

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

# Saving the model:

```python
joblib.dump(model, "penguin_clf.pkl")
```

# Testing the saved model:

#### Load the model and predict:

```python
penguin_clf = joblib.load("penguin_clf.pkl")
penguin_clf.predict(X_test)
```
